```python
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.drop(["ID"], axis=1, inplace=True)
```

```python
df['class'] = df['class'].replace({'P':1,'H':0})
df.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
x = df.drop(['class'], axis=1)
y = df['class']
x_centered = x - x.mean(axis=0)
```

```python
pca=PCA()
pca.fit(x_centered)
csum=np.cumsum(pca.explained_variance_ratio_)
d=np.argmax(csum>=0.99999)+1
```

```python
d
```

```python
pca=PCA(n_components=d)
X2D=pca.fit_transform(x_centered)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X2D, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
linear_svc = LinearSVC(C=0.7, loss="hinge", max_iter=100000)
linear_svc.fit(X_train_scaled, y_train)
```

```python
predictions = linear_svc.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```
